```python
lat_long_df = test_df[['start_lat','start_long']]
```

```python
# distance_matrix = squareform(pdist(lat_long_df, (lambda u,v: haversine(u,v))))

coords = lat_long_df.as_matrix(columns=['start_lat','start_long'])
```

```python
coords     #cluster coordinates (groups of more than 2 hours)
```

```python
kms_per_radian = 6371.0088
epsilon = 2.0 / kms_per_radian
# db = DBSCAN(eps=epsilon, min_samples=10, algorithm='ball_tree', metric='haversine').fit(np.radians(coords))
# cluster_labels = db.labels_
# num_clusters = len(set(cluster_labels))
# clusters = pd.Series([coords[cluster_labels == n] for n in range(num_clusters)])


# min_samples = 10 means 10 instances that any truck has stopped in that cluster 
#eps is the distance epsilon (here 2 km)



db = DBSCAN(eps=epsilon, min_samples=10, algorithm='ball_tree', metric='haversine').fit(np.radians(coords))

cluster_labels = db.labels_
n_clusters = len(set(cluster_labels))

# get the cluster
# cluster_labels = -1 means outliers
clusters = pd.Series([coords[cluster_labels == n] for n in range(-1, n_clusters)])
    
print('Number of clusters: {}'.format(n_clusters))
```

```python
add_lis = []  #list to store the centroid lat long
```

```python
for i in range(0,n_clusters):
    a = clusters[i]
    
    sum_lat = 0.0
    num_lat = 0
    sum_lon = 0.0
    num_lon = 0
    for j in range(0, len(a)):
        sum_lat += clusters[i][j][0]
        num_lat += 1
        sum_lon += clusters[i][j][1]
        num_lon += 1
    lis = [sum_lat/num_lat, sum_lon/num_lon, num_lat]
    add_lis.append(lis)
#     cluster_lat_long_df = cluster_lat_long_df.append(lis, ignore_index=True)
#     print lis
    
```

```python
cluster_lat_long_freq_df = pd.DataFrame(add_lis, columns=['lat','long','freq'])
```
